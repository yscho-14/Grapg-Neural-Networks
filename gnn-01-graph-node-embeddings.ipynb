```python
# Breadth-First Search (BFS)

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])

plt.axis('off')
nx.draw_networkx(G,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=600,
                 cmap='coolwarm',
                 font_size=14,
                 font_color='white'
                 )

def bfs(graph, node):
    visited, queue = [node], [node]

    while queue:
        node = queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

    return visited

bfs(G, 'A')
```

```python
# Depth-First Search (DFS)

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])

plt.axis('off')
nx.draw_networkx(G,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=600,
                 cmap='coolwarm',
                 font_size=14,
                 font_color='white'
                 )

visited = []

def dfs(visited, graph, node):
    if node not in visited:
        visited.append(node)
        for neighbor in graph[node]:
            visited = dfs(visited, graph, neighbor)

    return visited

dfs(visited, G, 'A')
```

```python
# Skip-gram Model

import re 
np.random.seed(0)

CONTEXT_SIZE = 2

raw_text = """Nature has a remarkable ability to recover from disturbances. 
When a forest is cleared by a wildfire, the ecosystem does not simply vanish. 
Instead, it begins a slow and methodical process of regeneration. This phenomenon, 
known as ecological succession, demonstrates the inherent strength of biological systems.
Initially, pioneer species like hardy weeds and grasses take root in the nutrient-rich ash. 
These organisms stabilize the soil and create a hospitable environment for more complex plants. 
Over several decades, shrubs and small trees begin to dominate the landscape, 
eventually leading to the return of a dense forest canopy.
Environmental scientists emphasize that biodiversity plays a crucial role in this recovery. 
A diverse range of species ensures that if one plant fails, others can fill its niche. 
Protecting these natural habitats is essential for maintaining the planet's ecological balance 
and ensuring a sustainable future for all living beings.
"""

clean_text = re.sub(r'[^\w\s]', '', raw_text).lower()
text = clean_text.split()

# Create skipgrams
skipgrams = []
for i in range(CONTEXT_SIZE, len(text) - CONTEXT_SIZE):
    array = [text[j] for j in np.arange(i - CONTEXT_SIZE, i + CONTEXT_SIZE + 1) if j != i]
    skipgrams.append((text[i], array))

print(skipgrams[:10])
```

```python
# Unique words
vocab = set(text)
VOCAB_SIZE = len(vocab)
print(f"Length of vocabulary = {VOCAB_SIZE}")
```

```python
from gensim.models.word2vec import Word2Vec

# Create Word2Vec
model = Word2Vec([text],
                 sg=1,   # 1: Skip-gram, 0: CBOW
                 vector_size=5,
                 min_count=0,
                 window=2,
                 workers=1,
                 seed=0)

print(f'Shape of W_embed: {model.wv.vectors.shape}')

# Train model
model.train([text], total_examples=model.corpus_count, epochs=10)

# Print a word embedding
print('\nWord embedding =')
print(model.wv[0])
print(model.wv[10])
print(model.wv[111])
```

```python
# Numerical Verification of Word Similarity

from sklearn.decomposition import PCA

word_a = "forest"
word_b = "trees"
word_c = "recovery" 

# Calculate after verifying if the word is in the model
if word_a in model.wv and word_b in model.wv:
    sim_ab = model.wv.similarity(word_a, word_b)
    print(f"Similarity({word_a}, {word_b}): {sim_ab:.4f}")
    
    sim_ac = model.wv.similarity(word_a, word_c)
    print(f"Similarity({word_a}, {word_c}): {sim_ac:.4f}")

else:
    print("The word is not in the model.")

# Extract top 3 words most similar to 'nature'
target_word = "nature"
if target_word in model.wv:
    print(f"\nTop 3 words most similar to '{target_word}':")
    print(model.wv.most_similar(target_word, topn=3))
```

```python
# Embedding Space Visualization (Principal Component Analysis)

from sklearn.decomposition import PCA

# Extract word vectors and their corresponding keys from the model
words = list(model.wv.index_to_key)
vectors = model.wv[words]

# Reduce 10D vectors to 2D space using PCA 
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)

# Scatter Plotting
plt.figure(figsize=(12, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], c='skyblue', edgecolors='b')

# Annotate each data point with its corresponding word label
for i, word in enumerate(words):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]), xytext=(5, 2),
                 textcoords='offset points', ha='right', va='bottom', fontsize=9)

plt.title("Word2Vec Word Embedding Visualization (2D PCA)")
plt.grid(True)
plt.show()
```

```python
# Molecule Preparation and Graph Conversion

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from gensim.models.word2vec import Word2Vec
from sklearn.decomposition import PCA

# Aspirin
smiles = "CC(=O)OC1=CC=CC=C1C(=O)O" 
mol = Chem.MolFromSmiles(smiles)
mol = Chem.AddHs(mol) # including Hydrogen

# Function to convert NetworkX Graph from RDKit Molecule
def mol_to_nx(mol):
    G = nx.Graph()
    
    for atom in mol.GetAtoms():
        G.add_node(atom.GetIdx(), symbol=atom.GetSymbol())
        
    for bond in mol.GetBonds():
        G.add_edge(bond.GetBeginAtomIdx(), bond.GetEndAtomIdx())
        
    return G

G = mol_to_nx(mol)

# Visualization
plt.figure(figsize=(8, 8)) 
pos = nx.kamada_kawai_layout(G) 

labels = {n: f"{G.nodes[n]['symbol']}{n}" for n in G.nodes}

node_symbols = nx.get_node_attributes(G, 'symbol')
colors = ['salmon' if node_symbols[n] == 'O' else 'skyblue' if node_symbols[n] == 'C' else 'lightgray' for n in G.nodes]

nx.draw(G, pos, 
        with_labels=True, 
        labels=labels,       
        node_color=colors, 
        node_size=800,       
        font_size=10,       
        font_weight='bold',
        edge_color='gray')

plt.title("Graph Representation of Aspirin (Numbered Atoms)")
plt.show()
```

```python
# Random Walk Generation
def random_walk(graph, start_node, length):
    walk = [str(start_node)]  # Convert to str as Word2Vec accepts string input
    curr = start_node
    
    for _ in range(length - 1):
        neighbors = list(graph.neighbors(curr))
        if len(neighbors) > 0:
            next_node = np.random.choice(neighbors)
            walk.append(str(next_node))
            curr = next_node
        else:
            break
    return walk

# random walks 80 times starting from every atom
walks = []
num_walks = 80
walk_length = 10

for node in G.nodes:
    for _ in range(num_walks):
        walks.append(random_walk(G, node, walk_length))

print(f"Generated {len(walks)} random walks.")
print(f"Sample walk from node 0: {walks[0]}")
print(f"Sample walk from node 1: {walks[80]}")
```

```python
# Atom Embedding Learning (Word2Vec)
# window=3: Treat atoms within 3 hops (chemical bonds)
model = Word2Vec(walks, vector_size=16, window=3, min_count=0, sg=1, workers=1, seed=42)
model.train(walks, total_examples=model.corpus_count, epochs=50)
```

```python
# Result Visualization (PCA)

# Extract vectors for each atom
node_indices = [str(i) for i in range(mol.GetNumAtoms())]
vectors = np.array([model.wv[i] for i in node_indices])

# Reduce dimensions to 2D using PCA
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)

# Color mapping by atomic symbol
symbols = [mol.GetAtomWithIdx(int(i)).GetSymbol() for i in node_indices]
color_map = {'C': 'skyblue', 'O': 'salmon', 'H': 'lightgray'}
node_colors = [color_map[s] for s in symbols]

# Scxatter plot
plt.figure(figsize=(10, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], c=node_colors, s=100, edgecolor='k', alpha=0.8)

# Annotate atom index and symbol on each point
for i, (x, y) in enumerate(vectors_2d):
    plt.text(x+0.02, y+0.02, f"{symbols[i]}{i}", fontsize=10)

# Add legend (Manual method)
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Carbon', markerfacecolor='skyblue', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='Oxygen', markerfacecolor='salmon', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='Hydrogen', markerfacecolor='lightgray', markersize=10)]
plt.legend(handles=legend_elements)

plt.title("Atom Embeddings of Aspirin (PCA Visualization)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```
